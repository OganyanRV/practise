```python
import numpy as np
import math
import matplotlib.pyplot as plt
def generate1():
    a = np.random.uniform(0, 1)
    b = np.random.uniform(0, 1)
    return (a * math.cos(2 * math.pi * b), a * math.sin(2 * math.pi * b))

def generate2():
    while True:
        x = np.random.uniform(-1, 1)
        y = np.random.uniform(-1, 1)
        if x ** 2 + y ** 2 > 1:
            continue
        return (x, y)
```

```python
gen1 = [generate1() for _ in range(1000)]
gen2 = [generate2() for _ in range(1000)]

gen1x, gen1y = list(zip(*gen1))
gen2x, gen2y = list(zip(*gen2))
```

```python
_, axes = plt.subplots(1,2, figsize=(20,10))
axes[0].scatter(gen1x, gen1y)
axes[1].scatter(gen2x, gen2y)
```

```python
ratio_gen1 = 0
for x,y in zip(gen1x, gen1y):
    if x <= 0.25 and x >= -0.25 and y>= -0.25 and y <= 0.25:
        ratio_gen1 += 1

ratio_gen1/= 1000
ratio_gen1
```

```python
ratio_gen2 = 0
for x,y in zip(gen2x, gen2y):
    if x <= 0.25 and x >= -0.25 and y>= -0.25 and y <= 0.25:
        ratio_gen2 += 1

ratio_gen2/= 1000
ratio_gen2
```
